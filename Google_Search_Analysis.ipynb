 Google search analysis can help a business understand what people are looking for at a particular point in time.

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt

import matplotlib.ticker as mtick
plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
google_search = TrendReq()
```

#Data Viz
We just going to analyze one or two search trends

```python
google_search.build_payload(kw_list=["Artificial Intelligence"])
data = google_search.interest_by_region()
data = data.sort_values(by="Artificial Intelligence", ascending=False)
data = data.head(20)
print(data)
```

From the above, the search queries on “Artificial Intelligence” are mostly done in Ethopia.

```python
#Let's visualize it some

data.reset_index().plot(x="geoName", y="Artificial Intelligence", figsize=(15,10), kind="bar")
                       
plt.show()
```

Let's visualise what the search querieshas been like in the last 5 years or so

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Artificial Intelligence'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 10))
data['Artificial Intelligence'].plot()
plt.title('Total Google Searches for Artificial Intelligence', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()#Greater interest in AI during 2018. it's been kind of steady since 2019
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 10))
data['Machine Learning'].plot()
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()#Searches on “machine learning” on Google started to increase in 2017 and the highest searches were done in 2020
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Bitcoin'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 10))
data['Bitcoin'].plot()
plt.title('Total Google Searches for Bitcoin', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()#2017 saw the increase of 'Bitcoin' searches and it peaked in 2018,suffered /
#the following two years and picked up great search interest again in 2021
```
